```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Untitled spreadsheet - Sheet1.csv", index_col=0)
```

```python
df = df.notna()
```

```python
df = df.transpose()
```

```python
df.head()
```

```python
counts_df = df.apply(pd.Series.value_counts)
```

```python
counts_df = counts_df.transpose()
```

```python
counts_df.columns
```

```python
counts_df = counts_df.drop(columns=[False])
```

```python
counts_df.sort_values(by=[True], inplace=True)
```

```python
counts_df.columns = ['Number of times topic has appeared in the last seven exams']
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)
```

```python
counts_df.plot.barh(figsize=(6,32), legend=False)
plt.savefig('dataminingexams.pdf')
```
